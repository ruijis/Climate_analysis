```python
import pandas as pd
```

```python
df = pd.read_csv('df_Richmond Point Potrero_USA_Clima.csv')
```

```python
df_summer = df[(df.month >= 6) & (df.month <= 8)]
df_summer
```

```python
df_winter = df[(df.month >= 12) | (df.month <= 2)]
df_winter
```

# Average parameters

```python
c = ['DBT', 'RH', 'dif_hor_rad', 'wind_speed', 'tot_sky_cover', 'Pwater']
for i in c:
    print(i)
    print(df_summer[i].mean())
```

```python
for i in c:
    print(i)
    print(df_winter[i].mean())
```

# Daily temp range

```python
df_max = df_summer.groupby(['month', 'day']).max()
df_min = df_summer.groupby(['month', 'day']).min()
df_range = df_max['DBT'] - df_min['DBT']
df_range.mean()
```

```python
df_max = df_winter.groupby(['month', 'day']).max()
df_min = df_winter.groupby(['month', 'day']).min()
df_range = df_max['DBT'] - df_min['DBT']
df_range.mean()
```
